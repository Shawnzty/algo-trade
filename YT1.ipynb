```python
aapl = close.AAPL.copy().to_frame() # convert to dataframe
aapl
```

```python
aapl["lag1"] = aapl.shift(periods=1) # shift by 1 period
aapl
```

```python
aapl["Diff"] = aapl.AAPL.sub(aapl.lag1)
aapl
```

```python
aapl["% Change"] = aapl.AAPL.div(aapl.lag1).sub(1).mul(100)
aapl
```

```python
aapl["Diff2"] = aapl.AAPL.diff(periods=1)
aapl
```

```python
aapl["% Change 2"] = aapl.AAPL.pct_change(periods=1).mul(100) # same as above
aapl
```

```python
del aapl["Diff"]
del aapl["Diff2"]
del aapl["lag1"]
del aapl["% Change"]
aapl
```

```python
aapl.rename(columns={"% Change 2":"Change"}, inplace=True)
aapl
```

```python
aapl.AAPL.resample("M").last() # last value of each month
```
