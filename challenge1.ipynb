```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
ticker = ["AAPL","KO","IBM","DIS","MSFT","GOOGL","ASML","TSLA","NVDA","AMD","VZ","T","GE","F","TM","ACRE","EPD","MPLX","BABA","BIDU"]
stocks = yf.download(ticker, start="2013-01-01", end="2023-01-01")
```

```python
close = stocks.loc[:,"Close"].copy()
```

```python
normclose = close.div(close.iloc[0]).mul(100)
```

```python
normclose.plot(figsize=(15,8), fontsize=12)
plt.legend(fontsize=12)
plt.show()
```

```python
ret = close.pct_change().dropna()
ret.head()
```

```python
ret.describe().T # Transpose
```

```python
summary = ret.describe().T.loc[:,["mean","std"]]
summary
```

```python
summary["mean"] = summary["mean"]*252
summary["std"] = summary["std"]*np.sqrt(252)
summary
```

```python
summary.plot.scatter(x="std", y="mean", figsize=(12,8), s=50, fontsize=15)
for i in summary.index:
    plt.annotate(i, (summary.loc[i,"std"]-0.01, summary.loc[i,"mean"]+0.01), fontsize=15)
plt.xlabel("Annual risk(std)", fontsize=15)
plt.ylabel("Annual return(mean)", fontsize=15)
plt.title("Risk/return", fontsize=25)
plt.show()
```
