```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
def get_sma(raw):
    data = raw.Close.to_frame()
    data["SMA50"] = data.rolling(window=50,min_periods=50).mean()
    data["SMA200"] = data.Close.rolling(window=200,min_periods=200).mean()
    data.plot(figsize=(12,8), fontsize=15)
    plt.legend(loc="upper left", fontsize=15)
    plt.show()
```

```python
get_sma(yf.download("AAPL", start="2013-01-01", end="2023-01-01"))
```
